# Exploratory Data Analysis
---

This section aims to investigate the global trends , upper middle income countries's trend, trends in Thailand , and how death rate in Thailand changes compared to upper middle income countries.

## Global Trends
---

```python
df_global = df[df['code'].notnull()].groupby('year').sum().reset_index()
```

```python
feature = ['outdoor_air_pollution',
       'high_systolic_blood_pressure', 'diet_high_in_sodium_',
       'diet_low_in_whole_grains', 'alochol_use', 'diet_low_in_fruits',
       'unsafe_water_source', 'secondhand_smoke', 'low_birth_weight',
       'child_wasting', 'unsafe_sex', 'diet_low_in_nuts_and_seeds',
       'household_air_pollution_from_solid_fuels', 'diet_low_in_vegetables',
       'low_physical_activity', 'smoking', 'high_fasting_plasma_glucose',
       'air_pollution', 'high_body_mass_index', 'unsafe_sanitation',
       'no_access_to_handwashing_facility', 'drug_use',
       'low_bone_mineral_density', 'vitamin_a_deficiency', 'child_stunting',
       'discontinued_breastfeeding', 'non-exclusive_breastfeeding',
       'iron_deficiency']

fig = plt.figure(figsize = (20,10))
fig.suptitle('Global Trends', fontsize=20 , style = 'italic')

for i,j in enumerate(feature):
    ax = fig.add_subplot(4,7,i+1)
    ax = sns.lineplot(df_global ,x = 'year',y = j)
    ax.set_title(j , fontsize = 14 , style = 'italic')
    ax.set_ylabel('Count')
plt.tight_layout()    
plt.show()  
```

## Comparing Changes in Death Rate between Thailand and Upper middle income countries
---

```python
df1 = df.copy()
```

```python
# df thailand
dfthai = df1[df1['entity'] == 'Thailand']

for i in feature:
    dfthai[i] = dfthai[i].pct_change()
    
dfthai = dfthai.fillna(0)
```

```python
# df upper middle income
dfmid = df1[df1['entity'] == 'World Bank Upper Middle Income']

for i in feature:
    dfmid[i] = dfmid[i].pct_change()
    
dfmid= dfmid.fillna(0)
```

```python
df_final = pd.concat([dfthai,dfmid] , axis = 0)
```

```python
fig = plt.figure(figsize = (24,14))
fig.suptitle('Thailand VS Upper Middle Income Countries', fontsize=20 , style = 'italic')
for i,j in enumerate(feature):
    ax = fig.add_subplot(4,7,i+1)
    ax = sns.lineplot(df_final ,x = 'year',y = j , hue = 'entity')
    ax.axhline(y  = 0 ,color="red", linestyle="--")
    ax.set_title(j , fontsize = 14 , style = 'italic')
    ax.set_ylabel('Percent Change')
    ax.get_legend().remove()

plt.legend()    
plt.tight_layout()    
plt.show()    
```
